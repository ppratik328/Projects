```python
df_num = df
```

```python
for col_name in df_num.columns:
    if(df_num[col_name].dtype == 'object' ):
        df_num[col_name] = df[col_name].astype('category')
        df_num[col_name] = df[col_name].cat.codes
```

```python
df_num
```

```python
df_num.dtypes
```

```python
CorMatrix = df_num.corr(method="pearson")
sns.heatmap(CorMatrix , annot=True)
plt.title("Correlation Matrix")
plt.xlable("Movie features")
plt.ylable("Movie features")
plt.show()
```
